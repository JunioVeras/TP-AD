```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

### Preparando dados

```python
DATA_PATH = "./data"
FILE_NAME = "stackoverflow.csv"
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv(f"{DATA_PATH}/{FILE_NAME}")
df.sample(5)
```

### AAA

```python
def filter_techs(row):
    techs = np.char.split(row["LanguageHaveWorkedWith"], sep=';').tolist()
    # techs = list(map(lambda str: str.strip(), techs))
    # techs = list(filter(not_desired_crimes, techs))
    # techs = map(cast_crime, techs)
    return techs

final_df = df[["ResponseId", "LanguageHaveWorkedWith"]]
final_df = final_df.dropna(subset=['LanguageHaveWorkedWith'])
final_df["items"] = final_df.apply(filter_techs, axis=1)
final_df
```

```python
techs_pre_explode = final_df["items"].to_numpy()
all_techs = np.unique(np.concatenate(techs_pre_explode))
all_techs, all_techs.shape
```

```python
TECHS_LEN = all_techs.shape[0]

final_map = {}
size = 1

for i in range(TECHS_LEN):
    final_map[all_techs[i]] = size
    size += 1
```

### Formato de entrada

```python
OUTPUT_PATH = "./out"
FILE_NAME = "techs.txt"
```

```python
with open(f"{OUTPUT_PATH}/{FILE_NAME}", 'w') as file:
    file.write("@CONVERTED_FROM_TEXT\n")
    # Header
    for (key, item) in final_map.items():
        file.write(f"@ITEM={item}={key}\n")
    # Sequences
    for row in final_df.itertuples():
        casted_items = map(lambda x: str(final_map[x]), row.items)
        file.write(f"{' '.join(casted_items)}\n")
```
